```python
import pandas as pd

data = pd.read_csv("bank-full.csv", sep=";")

data.head()
```

```python
# è³‡æ–™åˆ†æï¼šæ¬„ä½è³‡è¨Š
print("=== è³‡æ–™é›†åŸºæœ¬è³‡è¨Š ===")
print(f"è³‡æ–™ç­†æ•¸: {data.shape[0]:,}")
print(f"æ¬„ä½æ•¸é‡: {data.shape[1]}")
print(f"è³‡æ–™å¤§å°: {data.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print("\n=== æ¬„ä½åç¨±å’Œè³‡æ–™å‹æ…‹ ===")
data.info()

print("\n=== å„æ¬„ä½çš„å”¯ä¸€å€¼æ•¸é‡ ===")
for col in data.columns:
    unique_count = data[col].nunique()
    print(f"{col}: {unique_count} å€‹å”¯ä¸€å€¼")
    if unique_count <= 10:  # å¦‚æœå”¯ä¸€å€¼å°‘æ–¼ç­‰æ–¼10å€‹ï¼Œé¡¯ç¤ºæ‰€æœ‰å€¼
        print(f"  â†’ {sorted(data[col].unique())}")
    print()
```

```python
# (æ‰‹å‹•)ç§»é™¤ä¸å¿…è¦çš„ç‰¹å¾µ
# ç§»é™¤ dayå› ç‚ºèˆ‡ month ç›¸é—œï¼Œè€Œä¸”å“ªä¸€å¤©å½±éŸ¿ä¸å¤§
# ç§»é™¤ maritalã€educationã€contactã€pdays å› ç‚ºèˆ‡æ˜¯å¦è²¸æ¬¾è¼ƒç„¡é—œ
my_choice_data = data.drop(['day','marital','education','contact', 'pdays'],axis=1)
display(my_choice_data)
```

```python
# æª¢æŸ¥æœ‰ç„¡ç¼ºå¤±å€¼
print("=== æª¢æŸ¥ç¼ºå¤±å€¼ ===")
missing_values = my_choice_data.isnull().sum()
total_missing = missing_values.sum()
if total_missing == 0:
    print("âœ… è³‡æ–™é›†ä¸­æ²’æœ‰ç¼ºå¤±å€¼ã€‚")
else:
    print(f"âš ï¸ è³‡æ–™é›†ä¸­æœ‰ {total_missing} å€‹ç¼ºå¤±å€¼ï¼Œè«‹é€²è¡Œè™•ç†ã€‚")
    print(missing_values[missing_values > 0])
my_choice_data.isnull().sum()
```

```python
# è½‰æ›é¡åˆ¥å‹è®Šæ•¸ç‚ºæ•¸å€¼å‹è®Šæ•¸
# ä½¿ç”¨LabelEncoderå°‡è³‡æ–™é¡å‹è½‰æ›ç‚ºNumber
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in my_choice_data.select_dtypes(include=['object']).columns:
  my_choice_data[col] = le.fit_transform(my_choice_data[col])

display(my_choice_data)
```

```python
# # å°‡è³‡æ–™æ¡†åŒ¯å‡ºæˆCSVæª”æ¡ˆ
# print("=== åŒ¯å‡ºè³‡æ–™åˆ°CSVæª”æ¡ˆ ===")

# # æ–¹æ³•1ï¼šåŒ¯å‡ºåˆ°ç•¶å‰ç›®éŒ„
# my_choice_data.to_csv("bank_my_choice.csv", index=False)
# print("âœ… å·²åŒ¯å‡ºåˆ° bank_my_choice.csv")

# # æ–¹æ³•2ï¼šåŒ¯å‡ºåˆ°æŒ‡å®šè·¯å¾‘ï¼ˆå¯é¸ï¼‰
# # my_choice_data.to_csv("/path/to/your/bank_my_choice.csv", index=False)

# # æ–¹æ³•3ï¼šåŒ¯å‡ºæ™‚æŒ‡å®šç·¨ç¢¼ï¼ˆè™•ç†ä¸­æ–‡å­—ç¬¦ï¼‰
# # my_choice_data.to_csv("bank_my_choice_utf8.csv", index=False, encoding='utf-8')

# # æª¢æŸ¥åŒ¯å‡ºçš„æª”æ¡ˆè³‡è¨Š
# print(f"åŒ¯å‡ºè³‡æ–™ç¶­åº¦: {my_choice_data.shape}")
# print(f"åŒ¯å‡ºæ¬„ä½: {list(my_choice_data.columns)}")

# # é©—è­‰åŒ¯å‡ºçµæœ - è®€å–å‰›åŒ¯å‡ºçš„æª”æ¡ˆä¾†ç¢ºèª
# verify_data = pd.read_csv("bank_my_choice.csv")
# print(f"é©—è­‰è®€å–ç¶­åº¦: {verify_data.shape}")
# print("âœ… CSVæª”æ¡ˆåŒ¯å‡ºæˆåŠŸï¼")
```

```python
# è³‡æ–™åˆ†å‰² x å’Œ y
# æ³¨æ„ï¼šscalered_dataæ˜¯numpy arrayï¼Œéœ€è¦å…ˆåˆ†é›¢ç‰¹å¾µå’Œç›®æ¨™è®Šæ•¸å†é€²è¡Œæ­£è¦åŒ–

# æ–¹æ³•1ï¼šå¾åŸå§‹è³‡æ–™åˆ†é›¢ï¼Œç„¶å¾Œåˆ†åˆ¥è™•ç†
print("=== æ­£ç¢ºçš„è³‡æ–™åˆ†å‰²å’Œæ­£è¦åŒ–æµç¨‹ ===")

# å…ˆå¾åŸå§‹è³‡æ–™åˆ†é›¢ç‰¹å¾µå’Œç›®æ¨™è®Šæ•¸
X = my_choice_data.drop('y', axis=1)  # ç‰¹å¾µ
y = my_choice_data['y']               # ç›®æ¨™è®Šæ•¸

print(f"ç‰¹å¾µçŸ©é™£å½¢ç‹€: {X.shape}")
print(f"ç›®æ¨™è®Šæ•¸å½¢ç‹€: {y.shape}")

# é€²è¡Œè³‡æ–™åˆ†å‰²
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python

# æº–å‚™æ­£è¦åŒ–æ‰€éœ€çš„å¥—ä»¶
from sklearn.preprocessing import StandardScaler
# åªå°ç‰¹å¾µé€²è¡Œæ­£è¦åŒ–ï¼ˆç›®æ¨™è®Šæ•¸é€šå¸¸ä¸éœ€è¦æ­£è¦åŒ–ï¼‰
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # æ³¨æ„ï¼šæ¸¬è©¦é›†ä½¿ç”¨transformï¼Œä¸æ˜¯fit_transform

print(f"è¨“ç·´é›†ç‰¹å¾µå½¢ç‹€: {X_train_scaled.shape}")
print(f"æ¸¬è©¦é›†ç‰¹å¾µå½¢ç‹€: {X_test_scaled.shape}")
print(f"è¨“ç·´é›†ç›®æ¨™å½¢ç‹€: {y_train.shape}")
print(f"æ¸¬è©¦é›†ç›®æ¨™å½¢ç‹€: {y_test.shape}")

print("\nâœ… è³‡æ–™åˆ†å‰²å’Œæ­£è¦åŒ–å®Œæˆï¼")
```

```python
# ğŸ“š æ©Ÿå™¨å­¸ç¿’è³‡æ–™è™•ç†çš„æ­£ç¢ºæµç¨‹èªªæ˜
print("=== æ©Ÿå™¨å­¸ç¿’è³‡æ–™è™•ç†çš„æ­£ç¢ºé †åº ===")
print()
print("âœ… æ­£ç¢ºæµç¨‹ï¼š")
print("1. è³‡æ–™æ¸…ç† (ç§»é™¤ç¼ºå¤±å€¼ã€ç•°å¸¸å€¼)")
print("2. ç‰¹å¾µå·¥ç¨‹ (ç·¨ç¢¼é¡åˆ¥è®Šæ•¸)")
print("3. åˆ†é›¢ç‰¹å¾µ(X)å’Œç›®æ¨™è®Šæ•¸(y)")
print("4. è³‡æ–™åˆ†å‰² (train/test split)")
print("5. ç‰¹å¾µæ­£è¦åŒ– (åªå°ç‰¹å¾µï¼Œä¸å°ç›®æ¨™è®Šæ•¸)")
print("6. æ¨¡å‹è¨“ç·´")
print()
print("âš ï¸ é‡è¦æ³¨æ„äº‹é …ï¼š")
print("â€¢ æ­£è¦åŒ–æ‡‰è©²åœ¨è³‡æ–™åˆ†å‰²ä¹‹å¾Œé€²è¡Œ")
print("â€¢ åªå°ç‰¹å¾µé€²è¡Œæ­£è¦åŒ–ï¼Œç›®æ¨™è®Šæ•¸é€šå¸¸ä¿æŒåŸæ¨£")
print("â€¢ æ¸¬è©¦é›†ä½¿ç”¨è¨“ç·´é›†çš„scaleråƒæ•¸(é¿å…è³‡æ–™æ´©éœ²)")
print("â€¢ StandardScalerè¼¸å‡ºæ˜¯numpy arrayï¼Œä¸æ˜¯DataFrame")

# æª¢è¦–æˆ‘å€‘è™•ç†å¾Œçš„è³‡æ–™
print(f"\nğŸ“Š ç›®å‰è³‡æ–™ç‹€æ…‹ï¼š")
print(f"X_train_scaled type: {type(X_train_scaled)}")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"y_train type: {type(y_train)}")
print(f"y_train shape: {y_train.shape}")

# é¡¯ç¤ºå‰å¹¾è¡Œè³‡æ–™ï¼ˆè½‰æ›ç‚ºDataFrameæ–¹ä¾¿æŸ¥çœ‹ï¼‰
print(f"\nå‰5è¡Œæ­£è¦åŒ–å¾Œçš„ç‰¹å¾µè³‡æ–™ï¼š")
feature_names = my_choice_data.drop('y', axis=1).columns
X_train_df = pd.DataFrame(X_train_scaled, columns=feature_names)
display(X_train_df.head())

print(f"\nå‰5è¡Œç›®æ¨™è®Šæ•¸ï¼š")
display(y_train.head())
```

```python
# ç·šæ€§è¿´æ­¸æ¨¡å‹è¨“ç·´ (ç·šæ€§)
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)

score = lr_model.score(X_test_scaled, y_test)

print(f"ç·šæ€§è¿´æ­¸æ“¬åˆæˆåº¦: {score}")
print(f"ä¿‚æ•¸: {lr_model.coef_}")
```

```python
# é‚è¼¯è¿´æ­¸æ¨¡å‹è¨“ç·´ (éç·šæ€§)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr_model = LogisticRegression()
lr_model.fit(X_train_scaled, y_train)
y_pred = lr_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"é‚è¼¯è¿´æ­¸æº–ç¢ºç‡: {accuracy}")
```
